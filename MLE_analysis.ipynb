```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset("data/output.nc").squeeze()
ds = ds.assign_coords(yC=ds.yC*1e-3,yF=ds.yF*1e-3)
ds = ds.assign_coords(time=ds.time.astype("float")*1e-9/86400)
```

```python
(-ds["h"]).sel(yC=slice(-200,200)).mean("yC").plot()
```

```python
(-ds["h"]).sel(yC=slice(-200,200)).plot()
```

```python
ti = -1
fig,ax = plt.subplots(2, 1, figsize=(8,8))

fig.subplots_adjust(wspace=0.5)

ds["by"].isel(time=ti).plot.contourf(ax=ax[0], cmap="RdBu_r", vmin=-1e-7, vmax=1e-7, levels=12)
ds["x"].isel(time=ti).plot.contourf(ax=ax[1], cmap="PiYG_r", vmin=-3, vmax=3, levels=20)

for a in ax:
    (-ds["h"]).isel(time=ti).plot(ax=a, color="red")
    ds["b"].isel(time=ti).plot.contour(ax=a, levels=20, colors="k", linestyles="-")
    a.set(
        ylim=[-250,0],
        xlim=[-200,200],
        xlabel="yC [km]",
        title="",
    )
```

```python
ti = 1
fig,ax = plt.subplots(2, 1, figsize=(8,8))

fig.subplots_adjust(wspace=0.5)

ds["by"].isel(time=ti).plot.contourf(ax=ax[0], cmap="RdBu_r", vmin=-1e-7, vmax=1e-7, levels=12)
ds["x"].isel(time=ti).plot.contourf(ax=ax[1], cmap="PiYG_r", levels=11)

for a in ax:
    (-ds["h"]).isel(time=ti).plot(ax=a, color="red")
    ds["b"].isel(time=ti).plot.contour(ax=a, levels=20, colors="k", linestyles="-")
    a.set(
        ylim=[-250,0],
        xlim=[-200,200],
        xlabel="yC [km]",
        title="",
    )
```
