```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
dsy = xr.open_dataset("data/output.nc")
# dsy = dsy.interp(yF=dsy.yC, zF=dsy.zC)
dsy = dsy.assign_coords(time=dsy.time.astype("float")*1e-9/86400)

ti = 15

fig,ax = plt.subplots(2, 1, figsize=(8,5));

for a in ax:
    dsy.isel(time=ti).b.squeeze().plot.contour(ax=a, levels=20, colors="0.2", linestyles="-")
    (-dsy.isel(time=ti).h.squeeze()).plot(ax=a, color="green", marker="o", alpha=0.5)
    (-dsy.isel(time=ti).hx.squeeze()).plot(ax=a, color="red", marker="o", alpha=0.5)
    dsy.isel(time=ti).x.squeeze().plot.contour(levels=10, colors="0.2", ax=a)

dsy.isel(time=ti).v_mle.plot(ax=ax[0])
dsy.isel(time=ti).w_mle.plot(ax=ax[1])
_ = [a.set(title="", xlim=[-170e3, 170e3], ylim=[-200, 0]) for a in ax]
```
