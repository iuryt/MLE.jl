```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset("data/output.nc").squeeze()

ti = 0
fig,ax = plt.subplots(1, 2, figsize=(10,4))

fig.subplots_adjust(wspace=0.5)

ds["b"].isel(time=0).plot.contourf(ax=ax[0], cmap="RdBu_r", vmin=-1e-7, vmax=1e-7)
ds[""].isel(time=ti).plot.contourf(ax=ax[1], cmap="PiYG_r", vmin=-12, vmax= 12, levels=10)

for a in ax:
    (-ds["h"]).isel(time=0).plot(ax=a, color="red")
    ds["b"].isel(time=ti).plot.contour(ax=a, levels=20, colors="k", linestyles="-")
    a.set(
        ylim=[-400,0],
        title="",
    )



```
